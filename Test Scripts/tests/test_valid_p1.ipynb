```python
import json
```

```python
class InputAction:
    def __init__(self, actionType, properties):
        self.ActionType = actionType
        self.Properties = properties
        
        
    def toJSON(self):
        return json.dumps(self, default=lambda obj: obj.__dict__)
```

```python
class Action:
    def __init__(self, x = 0, y = 0, cardId = 0, amount = 0):
        self.X = x
        self.Y = y
        self.CardId = cardId
        self.Amount = amount
```
